1.Importing required libraries

```python
"""Movie Genre Classifier

This script performs movie genre classification using machine learning models. It includes data loading, visualization, model training, and predictions.

"""

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
```

2.Loading the datasets

```python
# Load the datasets
def load_data(filepath):
    """Load data from a CSV file with ':::' separator."""
    return pd.read_csv(filepath, sep=':::', names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])

train_data = load_data("/content/train_data.txt")
test_data = load_data("/content/test_data.txt")
test_solution_data = load_data("/content/test_data_solution.txt")
```

Visualizing our data

```python
# Display the first few rows and shape of each dataset
print("Training Data:")
display(train_data.head())
print(f"Shape: {train_data.shape}")

print("Test Data:")
display(test_data.head())
print(f"Shape: {test_data.shape}")

print("Test Solution Data:")
display(test_solution_data.head())
print(f"Shape: {test_solution_data.shape}")
```

```python
# 2. Visualize the data
# Plot distribution of movie genres
plt.figure(figsize=(20, 8))
sns.countplot(y=train_data['GENRE'], order=train_data['GENRE'].value_counts().index)
plt.title('Number of Movies per Genre')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```

```python
# Plot description length by genre
train_data['DESCRIPTION_length'] = train_data['DESCRIPTION'].apply(len)
plt.figure(figsize=(15, 10))
sns.barplot(x='GENRE', y='DESCRIPTION_length', data=train_data)
plt.title('Description Length by Genre')
plt.xticks(rotation=45)
plt.xlabel('Genre')
plt.ylabel('Description Length')
plt.show()

```

```python
# Plot top 10 most frequent genres
top_genres = train_data['GENRE'].value_counts().head(10)
plt.figure(figsize=(20, 10))
top_genres.plot(kind='barh', color='cyan')
plt.title('Top 10 Most Frequent Genres')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 3. Train the model
# Fill missing values in descriptions
train_data['DESCRIPTION'].fillna("", inplace=True)
test_data['DESCRIPTION'].fillna("", inplace=True)

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=100000)

# Transform text data
X_train = tfidf_vectorizer.fit_transform(train_data['DESCRIPTION'])
X_test = tfidf_vectorizer.transform(test_data['DESCRIPTION'])
```

```python
# Encode labels
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_data['GENRE'])
y_test = label_encoder.transform(test_solution_data['GENRE'])

# Split training data into training and validation subsets
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Train Linear SVC model
svc_model = LinearSVC()
svc_model.fit(X_train_sub, y_train_sub)
```

```python
# Evaluate model on validation data
y_val_pred = svc_model.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_val_pred))
print("Validation Classification Report:\n", classification_report(y_val, y_val_pred))
```

```python
# Predict on test data
y_pred = svc_model.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Test Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Train and evaluate Naive Bayes model
mnb_classifier = MultinomialNB()
mnb_classifier.fit(X_train, y_train)
print("Naive Bayes Test Predictions:", mnb_classifier.predict(X_test))
```

Predicting Movie Genres

```python
# Train and evaluate Logistic Regression model
lr_classifier = LogisticRegression(max_iter=500)
lr_classifier.fit(X_train, y_train)
print("Logistic Regression Test Predictions:", lr_classifier.predict(X_test))
```

```python
# Function to predict movie genre based on description
def predict_movie(description, model=svc_model, vectorizer=tfidf_vectorizer, encoder=label_encoder):
    """
    Predict the genre of a movie based on its description.

    Args:
        description (str): Movie description.
        model: Trained classification model (default is LinearSVC).
        vectorizer: TF-IDF vectorizer used for text transformation (default is tfidf_vectorizer).
        encoder: Label encoder used for encoding/decoding genre labels (default is label_encoder).

    Returns:
        str: Predicted movie genre.
    """
    description_vector = vectorizer.transform([description])
    pred_label = model.predict(description_vector)
    return encoder.inverse_transform(pred_label)[0]

# Example predictions
sample_descr_for_movie = "A movie where police catches the criminal and shoots him."
print("Predicted Genre:", predict_movie(sample_descr_for_movie))
```
